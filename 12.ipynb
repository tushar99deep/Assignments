```python

#**Question 1**
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


def print_linked_list(head):
    current = head
    while current is not None:
        print(current.value, end=" ")
        current = current.next
    print()

def delete_middle_node(head):
    if head is None or head.next is None:
        return head

    slow = head
    fast = head
    prev = None

    while fast is not None and fast.next is not None:
        prev = slow
        slow = slow.next
        fast = fast.next.next

    prev.next = slow.next

    return head
# Create a linked list: 1->2->3->4->5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

print_linked_list(head)  

head = delete_middle_node(head)

print_linked_list(head)  
```

```python
#**Question 2**


def has_loop(head):
    if head is None or head.next is None:
        return False

    slow = head
    fast = head.next

    while slow != fast:
        if fast is None or fast.next is None:
            return False

        slow = slow.next
        fast = fast.next.next

    return True

# Create a linked list with a loop: 1->2->3->4->5->3 (loop)
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = head.next.next

result = has_loop(head)
print(result) 

# Create a linked list without a loop: 1->2->3->4->5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

result = has_loop(head)
print(result) 
```

```python
#**Question 4**


def is_palindrome(head):
    if head is None or head.next is None:
        return True

    slow = head
    fast = head

    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

    # Reverse the second half of the linked list
    prev = None
    curr = slow
    while curr is not None:
        next_node = curr.next
        curr.next = prev
        prev = curr
        curr = next_node

    # Compare the first half with the reversed second half
    p1 = head
    p2 = prev

    while p1 is not None and p2 is not None:
        if p1.value != p2.value:
            return False
        p1 = p1.next
        p2 = p2.next

    return True

# Create a palindrome linked list: R->A->D->A->R
head = Node('R')
head.next = Node('A')
head.next.next = Node('D')
head.next.next.next = Node('A')
head.next.next.next.next = Node('R')

result = is_palindrome(head)
print(result)  

# Create a non-palindrome linked list: C->O->D->E
head = Node('C')
head.next = Node('O')
head.next.next = Node('D')
head.next.next.next = Node('E')

result = is_palindrome(head)
print(result)
```

```python
#**Question 5**


def remove_loop(head):
    if head is None or head.next is None:
        return head

    slow = head
    fast = head

    # Check if there is a loop
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break

    # If there is a loop, find the start of the loop
    if slow == fast:
        slow = head
        while slow.next != fast.next:
            slow = slow.next
            fast = fast.next

        # Remove the loop
        fast.next = None

    return head

# Create a linked list with a loop: 1->3->4->3 (loop)
head = Node(1)
head.next = Node(3)
head.next.next = Node(4)
head.next.next.next = head.next

head = remove_loop(head)

print_linked_list(head)  # Output: 1->3->4

# Create a linked list without a loop: 1->2->3->4->5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

head = remove_loop(head)

print_linked_list(head) 
```

```python
#**Question 6**



def retain_delete(head, M, N):
    if head is None or M <= 0 or N <= 0:
        return head

    current = head

    while current is not None:
        # Traverse M nodes
        for _ in range(M-1):
            if current.next is None:
                return head
            current = current.next

        # Delete N nodes
        temp = current.next
        for _ in range(N):
            if temp is None:
                break
            temp = temp.next
        current.next = temp
        current = temp

    return head





# Create a linked list: 1->2->3->4->5->6->7->8
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

head = retain_delete(head, 2, 2)

print_linked_list(head) 
```

```python

'''
 **Question 8**

Given a singly linked list, find if the linked list is [circular](https://www.geeksforgeeks.org/circular-linked-list/amp/) 
or not.

> A linked list is called circular if it is not NULL-terminated and all nodes are connected in the form of a cycle. 
Below is an example of a circular linked list.
> 
</aside>'''

def is_circular(head):
    if head is None:
        return False

    slow = head
    fast = head.next

    while fast is not None and fast.next is not None:
        if slow == fast:
            return True

        slow = slow.next
        fast = fast.next.next

    return False





# Creating a circular linked list: 1->2->3->4->5->3 (loop)
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = head.next.next

result = is_circular(head)
print(result)  # Output: True

# Creating a non-circular linked list: 1->2->3->4->5
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)

result = is_circular(head)
print(result) 


```
