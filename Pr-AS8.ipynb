```python
def add_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    return result

# Matrix 1
rows = int(input("Enter the number of rows for matrix 1: "))
cols = int(input("Enter the number of columns for matrix 1: "))

matrix1 = []
print("Enter the elements of matrix 1:")
for i in range(rows):
    row = [int(x) for x in input().split()]
    matrix1.append(row)

# Matrix 2
print("Enter the elements of matrix 2:")
matrix2 = []
for i in range(rows):
    row = [int(x) for x in input().split()]
    matrix2.append(row)

# Addition
sum_matrix = add_matrices(matrix1, matrix2)

# Displaying the result
print("Matrix 1:")
for row in matrix1:
    print(row)

print("Matrix 2:")
for row in matrix2:
    print(row)

print("Sum of the matrices:")
for row in sum_matrix:
    print(row)
```

```python
def multiply_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            element = 0
            for k in range(len(matrix2)):
                element += matrix1[i][k] * matrix2[k][j]
            row.append(element)
        result.append(row)
    return result

# Matrix 1
rows1 = int(input("Enter the number of rows for matrix 1: "))
cols1 = int(input("Enter the number of columns for matrix 1: "))

matrix1 = []
print("Enter the elements of matrix 1:")
for i in range(rows1):
    row = [int(x) for x in input().split()]
    matrix1.append(row)

# Matrix 2
rows2 = int(input("Enter the number of rows for matrix 2: "))
cols2 = int(input("Enter the number of columns for matrix 2: "))

matrix2 = []
print("Enter the elements of matrix 2:")
for i in range(rows2):
    row = [int(x) for x in input().split()]
    matrix2.append(row)

# Multiplication
product_matrix = multiply_matrices(matrix1, matrix2)

# Displaying the result
print("Matrix 1:")
for row in matrix1:
    print(row)

print("Matrix 2:")
for row in matrix2:
    print(row)

print("Product of the matrices:")
for row in product_matrix:
    print(row)
```

```python
def transpose_matrix(matrix):
    transposed = [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]
    return transposed

# Matrix
rows = int(input("Enter the number of rows: "))
cols = int(input("Enter the number of columns: "))

matrix = []
print("Enter the elements of the matrix:")
for i in range(rows):
    row = [int(x) for x in input().split()]
    matrix.append(row)

# Transpose
transposed_matrix = transpose_matrix(matrix)

# Displaying the result
print("Original Matrix:")
for row in matrix:
    print(row)

print("Transposed Matrix:")
for row in transposed_matrix:
    print(row)
```

```python
sentence = input("Enter a sentence: ")

words = sentence.split()

sorted_words = sorted(words)

print("Sorted words in alphabetical order:")
for word in sorted_words:
    print(word)
```

```python
import string

def remove_punctuation(text):
    translation_table = str.maketrans("", "", string.punctuation)
    stripped_text = text.translate(translation_table)
    return stripped_text

sentence = input("Enter a sentence: ")

stripped_sentence = remove_punctuation(sentence)

print("Sentence without punctuation:", stripped_sentence)
```
