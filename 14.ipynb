```python
# Question 2
class Node:
    def __init__(self, value):
        self.data = value
        self.next = None

def reverseLinkedList(head):
    prev = None
    current = head

    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return prev

def addOne(head):
    # Reverse the linked list
    head = reverseLinkedList(head)

    # Add 1 to the reversed number
    current = head
    carry = 1

    while current:
        sum = current.data + carry
        current.data = sum % 10
        carry = sum // 10

        if not current.next and carry:
            current.next = Node(carry)
            break

        current = current.next

    # Reverse the linked list back
    head = reverseLinkedList(head)

    return head
# Create a linked list
head = Node(4)
node2 = Node(5)
node3 = Node(6)

head.next = node2
node2.next = node3

# Call the addOne function
head = addOne(head)

# Print the modified linked list
current = head
while current:
    print(current.data, end=" ")
    current = current.next
```

```python
#Question 3

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
        self.bottom = None

def mergeTwoLists(list1, list2):
    dummy = Node(0)
    tail = dummy

    while list1 and list2:
        if list1.data < list2.data:
            tail.bottom = list1
            list1 = list1.bottom
        else:
            tail.bottom = list2
            list2 = list2.bottom

        tail = tail.bottom

    if list1:
        tail.bottom = list1
    else:
        tail.bottom = list2

    return dummy.bottom

def flattenLinkedList(head):
    if not head or not head.next:
        return head

    head.next = flattenLinkedList(head.next)

    head = mergeTwoLists(head, head.next)

    return head

# Create the linked list
head = Node(5)
head.next = Node(10)
head.next.next = Node(19)
head.next.next.next = Node(28)

head.bottom = Node(7)
head.bottom.bottom = Node(8)
head.bottom.bottom.bottom = Node(30)

head.next.bottom = Node(20)

head.next.next.bottom = Node(22)
head.next.next.bottom.bottom = Node(50)

head.next.next.next.bottom = Node(35)
head.next.next.next.bottom.bottom = Node(40)
head.next.next.next.bottom.bottom.bottom = Node(45)

# Call the flattenLinkedList function
head = flattenLinkedList(head)

# Print the flattened linked list
current = head
while current:
    print(current.data, end=" ")
    current = current.bottom
```

```python
#Question 4

class Node:
    def __init__(self, value):
        self.data = value
        self.next = None
        self.arb = None

def copyRandomList(head):
    if not head:
        return None

    # Create a mapping of original nodes to their corresponding copied nodes
    node_map = {}

    # Create the copied nodes and set their values
    current = head
    while current:
        copied_node = Node(current.data)
        node_map[current] = copied_node
        current = current.next

    # Set the next and random pointers for the copied nodes
    current = head
    while current:
        copied_node = node_map[current]
        if current.next:
            copied_node.next = node_map[current.next]
        if current.arb:
            copied_node.arb = node_map[current.arb]
        current = current.next

    return node_map[head]
# Create the original linked list
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)

head.next = node2
head.arb = node2

node2.next = node3
node2.arb = node4

node3.next = node4
node3.arb = node2

# Call the copyRandomList function
copied_head = copyRandomList(head)

# Print the copied linked list
current = copied_head
while current:
    print("Node:", current.data)
    if current.next:
        print("Next Pointer:", current.next.data)
    else:
        print("Next Pointer: None")
    if current.arb:
        print("Random Pointer:", current.arb.data)
    else:
        print("Random Pointer: None")
    print("")

    current = current.next
```

```python
#Question 5
class Node:
    def __init__(self, value):
        self.val = value
        self.next = None

def oddEvenList(head):
    if not head or not head.next:
        return head

    odd_head = head
    even_head = head.next

    odd = odd_head
    even = even_head

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head

    return odd_head

# Example usage:
# Create the linked list
head = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
node5 = Node(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Call the oddEvenList function
head = oddEvenList(head)

# Print the reordered list
current = head
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
#Question 6

class Node:
    def __init__(self, value):
        self.val = value
        self.next = None

def leftShiftLinkedList(head, k):
    if not head or not head.next or k == 0:
        return head

    # Find the length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Adjust k to the range [0, length)
    k = k % length

    if k == 0:
        return head

    # Find the kth node from the end
    slow = head
    fast = head

    for _ in range(k):
        fast = fast.next

    while fast.next:
        slow = slow.next
        fast = fast.next

    new_head = slow.next
    slow.next = None
    fast.next = head

    return new_head

# Example usage:
# Create the linked list
head = Node(2)
node2 = Node(4)
node3 = Node(7)
node4 = Node(8)
node5 = Node(9)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Call the leftShiftLinkedList function
k = 3
head = leftShiftLinkedList(head, k)

# Print the shifted linked list
current = head
while current:
    print(current.val, end=" ")
    current = current.next
```

```python
#Question 7
class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None

def nextLargerNodes(head):
    if not head:
        return []

    stack = []
    values = []
    result = []

    # Traverse the linked list and store the values in a list
    while head:
        values.append(head.val)
        head = head.next

    # Iterate through the values list
    for val in values:
        # Compare the current value with the values in the stack
        while stack and val > values[stack[-1]]:
            result[stack.pop()] = val
        # Push the index of the current value into the stack
        stack.append(len(result))
        # Initialize the result for the current value as 0
        result.append(0)

    return result

# Example usage:
# Create the linked list
head = ListNode(2)
node2 = ListNode(7)
node3 = ListNode(4)
node4 = ListNode(3)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Call the nextLargerNodes function
result = nextLargerNodes(head)

# Print the result
print(result)
```

```python
#Question 8

class ListNode:
    def __init__(self, value):
        self.val = value
        self.next = None

def removeZeroSumSublists(head):
    # Create a dummy node to handle the case when the head itself is part of the zero-sum sublist
    dummy = ListNode(0)
    dummy.next = head

    prefix_sum = 0
    prefix_sums = {}
    current = dummy

    # Calculate the prefix sums and store them in a dictionary
    while current:
        prefix_sum += current.val
        if prefix_sum in prefix_sums:
            # If the prefix sum already exists in the dictionary, remove the sublist from the linked list
            prev = prefix_sums[prefix_sum].next
            prefix_sum_removed = prefix_sum + prev.val
            while prev != current:
                del prefix_sums[prefix_sum_removed]
                prev = prev.next
                prefix_sum_removed += prev.val
            prefix_sums[prefix_sum].next = current.next
        else:
            # If the prefix sum is encountered for the first time, add it to the dictionary
            prefix_sums[prefix_sum] = current
        current = current.next

    return dummy.next

# Example usage:
# Create the linked list
head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(-3)
node5 = ListNode(4)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Call the removeZeroSumSublists function
head = removeZeroSumSublists(head)

# Print the resulting linked list
current = head
while current:
    print(current.val, end=" ")
    current = current.next
```
