```python
'''
Question1
Create a function that takes a string and returns a string in which each character is repeated once.
Examples
double_char("String") ➞ "SSttrriinngg"

double_char("Hello World!") ➞ "HHeelllloo  WWoorrlldd!!"

double_char("1234!_ ") ➞ "11223344!!__  "'''

class StringManipulator:
    def double_char(self, txt):
        doubled_str = ""
        for char in txt:
            doubled_str += char * 2
        return doubled_str

# Usage:
str_manipulator = StringManipulator()
result = str_manipulator.double_char("String")
print(result)  # Output: SSttrriinngg
```

```python
'''
Question2
Create a function that reverses a boolean value and returns the string "boolean expected" if another variable type is given.
Examples
reverse(True) ➞ False

reverse(False) ➞ True

reverse(0) ➞ "boolean expected"

reverse(None) ➞ "boolean expected"
'''

class BooleanReverser:
    def reverse(self, value):
        if isinstance(value, bool):
            return not value
        else:
            return "boolean expected"

# Usage:
bool_reverser = BooleanReverser()
result = bool_reverser.reverse(True)
print(result)  # Output: False

```

```python
'''Question3
Create a function that returns the thickness (in meters) of a piece of paper after folding it n number of times. The paper starts off with a thickness of 0.5mm.
Examples
num_layers(1) ➞ "0.001m"
# Paper folded once is 1mm (equal to 0.001m)

num_layers(4) ➞ "0.008m"
# Paper folded 4 times is 8mm (equal to 0.008m)

num_layers(21) ➞ "1048.576m"
# Paper folded 21 times is 1048576mm (equal to 1048.576m)

'''

class PaperFolder:
    def num_layers(self, n):
        thickness = 0.5
        for _ in range(n):
            thickness *= 2
        return f"{thickness/1000}m"

# Usage:
paper = PaperFolder()
result = paper.num_layers(1)
print(result)  # Output: 0.001m
```

```python
'''
Question4
Create a function that takes a single string as argument and returns an ordered list containing the indices of all capital letters in the string.
Examples
index_of_caps("eDaBiT") ➞ [1, 3, 5]

index_of_caps("eQuINoX") ➞ [1, 3, 4, 6]

index_of_caps("determine") ➞ []

index_of_caps("STRIKE") ➞ [0, 1, 2, 3, 4, 5]

index_of_caps("sUn") ➞ [1]
'''

class CapitalLetterIndices:
    def index_of_caps(self, txt):
        indices = []
        for index, char in enumerate(txt):
            if char.isupper():
                indices.append(index)
        return indices

# Usage:
caps_indices = CapitalLetterIndices()
result = caps_indices.index_of_caps("eDaBiT")
print(result)  # Output: [1, 3, 5]
```

```python
'''
Question5
Using list comprehensions, create a function that finds all even numbers from 1 to the given number.
Examples
find_even_nums(8) ➞ [2, 4, 6, 8]

find_even_nums(4) ➞ [2, 4]

find_even_nums(2) ➞ [2]

'''

class EvenNumberFinder:
    def find_even_nums(self, n):
        return [num for num in range(1, n + 1) if num % 2 == 0]

# Usage:
even_finder = EvenNumberFinder()
result = even_finder.find_even_nums(8)
print(result)  # Output: [2, 4, 6, 8]
```
