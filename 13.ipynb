```python
### Question 1: Create a New Linked List from Two Linked Lists

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def create_new_linked_list(list1, list2):
    if not list1:
        return list2
    if not list2:
        return list1

    head = None
    tail = None

    while list1 and list2:
        if list1.data > list2.data:
            new_node = Node(list1.data)
            list1 = list1.next
        else:
            new_node = Node(list2.data)
            list2 = list2.next

        if not head:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = tail.next

    if list1:
        tail.next = list1
    if list2:
        tail.next = list2

    return head


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating linked list 1: 5->2->3->8
list1 = Node(5)
list1.next = Node(2)
list1.next.next = Node(3)
list1.next.next.next = Node(8)

# Creating linked list 2: 1->7->4->5
list2 = Node(1)
list2.next = Node(7)
list2.next.next = Node(4)
list2.next.next.next = Node(5)

# Create the new linked list
new_list = create_new_linked_list(list1, list2)

# Print the new linked list
print_linked_list(new_list)




```

```python
### Question 2: Remove Duplicates from a Sorted Linked List

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def remove_duplicates(head):
    if not head or not head.next:
        return head

    current = head

    while current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating the sorted linked list: 11->11->11->21->43->43->60
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicates
new_head = remove_duplicates(head)

# Print the modified linked list
print_linked_list(new_head)


```

```python
### Question 3: Reverse Linked List in Groups of k


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def reverse_k_nodes(head, k):
    if not head or k == 1:
        return head

    dummy = Node(0)
    dummy.next = head
    prev = dummy

    while head:
        tail = prev
        for _ in range(k):
            tail = tail.next
            if not tail:
                return dummy.next

        next_head = tail.next
        head, tail = reverse_linked_list(head, tail)
        prev.next = head
        tail.next = next_head
        prev = tail
        head = next_head

    return dummy.next


def reverse_linked_list(head, tail):
    prev = None
    current = head

    while prev != tail:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node

    return tail, head


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating the linked list: 1->2->2->4->5->6->7->8
head = Node(1)
head.next = Node(2)
head.next.next = Node(2)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

# Reverse the linked list in groups of 4
new_head = reverse_k_nodes(head, 4)

# Print the modified linked list
print_linked_list(new_head)



```

```python
### Question 4: Reverse Alternate k Nodes in a Linked List


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def reverse_alternate_k_nodes(head, k):
    if not head or k == 1:
        return head

    current = head
    prev = None
    count = 0

    while current and count < k:
        next_node = current.next
        if count % 2 == 0:
            current.next = prev
            prev = current
        else:
            current.next = next_node
        current = next_node
        count += 1

    if current:
        head.next = reverse_alternate_k_nodes(current, k)

    return prev


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating the linked list: 1->2->3->4->5->6->7->8->9->NULL
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

# Reverse alternate k nodes
new_head = reverse_alternate_k_nodes(head, 3)

# Print the modified linked list
print_linked_list(new_head)
```

```python
#Question 5


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def delete_last_occurrence(head, key):
    # Check if the linked list is empty
    if not head:
        return head

    # Initialize pointers
    prev = None
    curr = head
    last_occurrence = None
    prev_last_occurrence = None

    # Traverse the linked list
    while curr:
        if curr.val == key:
            last_occurrence = curr
            prev_last_occurrence = prev

        prev = curr
        curr = curr.next

    # Check if the key was found
    if last_occurrence:
        # Delete the last occurrence
        if prev_last_occurrence:
            prev_last_occurrence.next = last_occurrence.next
        else:
            head = head.next

    return head


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.val, end=" ")
        head = head.next
    print()


# Test the function
# Creating the linked list: 1 -> 2 -> 3 -> 5 -> 2 -> 10
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(5)
head.next.next.next.next = ListNode(2)
head.next.next.next.next.next = ListNode(10)

# Delete the last occurrence of key = 2
new_head = delete_last_occurrence(head, 2)

# Print the modified linked list
print_linked_list(new_head)
```

```python
### Question 6: Merge Two Sorted Linked Lists


class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def merge_sorted_lists(head1, head2):
    if not head1:
        return head2
    if not head2:
        return head1

    if head1.data < head2.data:
        head = head1
        head.next = merge_sorted_lists(head1.next, head2)
    else:
        head = head2
        head.next = merge_sorted_lists(head1, head2.next)

    return head


# Helper function to print the linked list
def print_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating the sorted linked list a: 5->10->15
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

# Creating the sorted linked list b: 2->3->20
b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

# Merge the two sorted linked lists
merged_list = merge_sorted_lists(a, b)

# Print the merged linked list
print_linked_list(merged_list)
```

```python
#Question 7
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def reverse_doubly_linked_list(head):
    if not head or not head.next:
        return head

    current = head
    prev = None

    while current:
        next_node = current.next
        current.next = prev
        current.prev = next_node
        prev = current
        current = next_node

    return prev


# Helper function to print the doubly linked list
def print_doubly_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()


# Test the function
# Creating the doubly linked list: 10->8->4->2
head = Node(10)
head.next = Node(8)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next
head.next.next.next = Node(2)
head.next.next.next.prev = head.next.next

# Reverse the doubly linked list
new_head = reverse_doubly_linked_list(head)

# Print the reversed doubly linked list
print_doubly_linked_list(new_head)
```

```python
#Question 8

class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


def delete_node(head, position):
    if not head:
        return None

    if position == 1:
        if head.next:
            head.next.prev = None
        return head.next

    current = head
    count = 1

    while current and count < position:
        current = current.next
        count += 1

    if not current:
        return head

    current.prev.next = current.next

    if current.next:
        current.next.prev = current.prev

    return head


# Helper function to print the doubly linked list
def print_doubly_linked_list(head):
    while head:
        print(head.data, end=" ")
        head = head.next
    print()



# Creating the doubly linked list: 1<->3<->4
head = Node(1)
head.next = Node(3)
head.next.prev = head
head.next.next = Node(4)
head.next.next.prev = head.next

# Delete the node at position 3
new_head = delete_node(head, 3)

# Print the modified doubly linked list
print_doubly_linked_list(new_head)
```
