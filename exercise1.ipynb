```python
#q4 : Create a class to implement all insert , update , delete operation for mysql
import mysql.connector

class DBConnection:
    def __init__(self, host, username, password, db_name):
        self.host = host
        self.username = username
        self.password = password
        self.db_name = db_name
        self.cnx = mysql.connector.connect(host=self.host, user=self.username, password=self.password, database=self.db_name)
        self.cursor = self.cnx.cursor()
        
    def close(self):
        self.cursor.close()
        self.cnx.close()

class DMLandDDL(DBConnection):
    def insert(self, table, values):
        query = f"INSERT INTO {table} VALUES {values}"
        self.cursor.execute(query)
        self.cnx.commit()
        print(f"Inserted {self.cursor.rowcount} row(s) into {table}")
        
    def update(self, table, set_values, where_condition):
        query = f"UPDATE {table} SET {set_values} WHERE {where_condition}"
        self.cursor.execute(query)
        self.cnx.commit()
        print(f"Updated {self.cursor.rowcount} row(s) in {table}")
        
    def delete(self, table, where_condition):
        query = f"DELETE FROM {table} WHERE {where_condition}"
        self.cursor.execute(query)
        self.cnx.commit()
        print(f"Deleted {self.cursor.rowcount} row(s) from {table}")
        
    def select(self, table, columns='*', where_condition=None):
        query = f"SELECT {columns} FROM {table}"
        if where_condition:
            query += f" WHERE {where_condition}"
        self.cursor.execute(query)
        result = self.cursor.fetchall()
        print(f"Fetched {len(result)} row(s) from {table}")
        return result
    
    

ops=DMLandDDL('localhost','root','root','exercise1')

ops.insert('data',('tushar','tushardeep3476@gmail.com'))
ops.select('data')
```

```python
import time
import tracemalloc

class Complexity:
    def __init__(self):
        pass
    
    @staticmethod
    def time_complexity(fn, *args):
        start_time = time.time()
        fn(*args)
        end_time = time.time()
        return end_time - start_time

    @staticmethod
    def space_complexity(fn, *args):
        tracemalloc.start()
        fn(*args)
        current, peak = tracemalloc.get_traced_memory()
        tracemalloc.stop()
        return peak
    
class test(Complexity):
    def sample_function(n):
        return [i**2 for i in range(n)]

c = Complexity()
t = c.time_complexity(sample_function, 1000)
print(f"Time complexity: {t}")
s = c.space_complexity(sample_function, 1000)
print(f"Space complexity: {s}")
```
