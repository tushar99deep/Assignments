```python
'''
Question 1
Create a function that takes a number as an argument and returns True or False depending on whether the number is symmetrical or not. A number is symmetrical when it is the same as its reverse.
Examples
is_symmetrical(7227) ➞ True

is_symmetrical(12567) ➞ False

is_symmetrical(44444444) ➞ True

is_symmetrical(9939) ➞ False

is_symmetrical(1112111) ➞ True
'''

class IsSymmetrical:
    @staticmethod
    def is_symmetrical(num):
        return str(num) == str(num)[::-1]


# Usage example
is_symmetrical = IsSymmetrical()
print(is_symmetrical.is_symmetrical(7227))
```

```python
'''
Question 2
Given a string of numbers separated by a comma and space, return the product of the numbers.
Examples
multiply_nums("2, 3") ➞ 6

multiply_nums("1, 2, 3, 4") ➞ 24

multiply_nums("54, 75, 453, 0") ➞ 0

multiply_nums("10, -2") ➞ -20
'''

class MultiplyNums:
    @staticmethod
    def multiply_nums(nums):
        num_list = [int(num) for num in nums.split(", ")]
        product = 1
        for num in num_list:
            product *= num
        return product


# Usage example
multiply_nums = MultiplyNums()
print(multiply_nums.multiply_nums("2, 3"))
```

```python
'''Question 3
Create a function that squares every digit of a number.
Examples
square_digits(9119) ➞ 811181

square_digits(2483) ➞ 416649

square_digits(3212) ➞ 9414
Notes
The function receives an integer and must return an integer.
'''

class SquareDigits:
    @staticmethod
    def square_digits(num):
        squared_digits = [int(digit) ** 2 for digit in str(num)]
        return int(''.join(map(str, squared_digits)))


# Usage example
square_digits = SquareDigits()
print(square_digits.square_digits(9119))
```

```python
'''
Question 4
Create a function that sorts a list and removes all duplicate items from it.
Examples
setify([1, 3, 3, 5, 5]) ➞ [1, 3, 5]

setify([4, 4, 4, 4]) ➞ [4]

setify([5, 7, 8, 9, 10, 15]) ➞ [5, 7, 8, 9, 10, 15]

setify([3, 3, 3, 2, 1]) ➞ [1, 2, 3]
'''

class Setify:
    @staticmethod
    def setify(lst):
        return sorted(list(set(lst)))


# Usage example
setify = Setify()
print(setify.setify([1, 3, 3, 5, 5]))
```

```python
'''
Question 5
Create a function that returns the mean of all digits.
Examples
mean(42) ➞ 3

mean(12345) ➞ 3

mean(666) ➞ 6
Notes
The mean of all digits is the sum of digits / how many digits there are (e.g. mean of digits in 512 is (5+1+2)/3(number of digits) = 8/3=2).
The mean will always be an integer.

'''


class Mean:
    @staticmethod
    def mean(num):
        digits = [int(digit) for digit in str(num)]
        return sum(digits) // len(digits)


# Usage example
mean = Mean()
print(mean.mean(42))
```
