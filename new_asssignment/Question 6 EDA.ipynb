# Get all Pokemons whose spawn time is less than 300 seconds.

```python
data['spawn_time_minute']= data['spawn_time'].apply(lambda x: int(x.split(':')[0]) if isinstance(x, str) else 0)
#df['spawn_minutes'] = df['spawn_time'].apply(lambda x: int(x.split(':')[0]))
```

```python
data['spawn_time_seconds']= data['spawn_time'].apply(lambda x: int(x.split(':')[1]) if isinstance(x, str) else 0)
```

```python
data['spawn_time_minute']= data['spawn_time_minute'] * 60
```

```python
data['total_spawntime_sec']=data['spawn_time_minute'].add(data['spawn_time_seconds'])
```
