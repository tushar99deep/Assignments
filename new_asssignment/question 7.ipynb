```python
# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = data[(data['nametype'] == 'Valid') & (data['year'].dt.year < 2000)]
```

```python
# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_2000[earth_meteorites_before_2000['year'].dt.year<1970][['reclat','reclong']]
```
