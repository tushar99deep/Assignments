```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import st
```

```python
df=pd.read_csv('statsq6.csv')
```

```python
df.head()
```

```python
# Plotting a histogram
plt.hist(df[' Blood Pressure After (mmHg)'] - df[' Blood Pressure Before (mmHg)'], bins='auto', alpha=0.7, rwidth=0.85)
plt.xlabel('Change in Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.title('Distribution of Change in Blood Pressure')
plt.show()
```

```python
import scipy.stats as stats
# Creating a Q-Q plot
plt.figure(figsize=(6, 4))
stats.probplot(df[' Blood Pressure After (mmHg)'] - df[' Blood Pressure Before (mmHg)'], dist="norm", plot=plt)
plt.title("Q-Q plot of Change in Blood Pressure")
plt.show()
'''The Q-Q plot compares the observed data against the expected values from a normal distribution. If the points lie approximately on a straight line, it suggests that the data follows a normal distribution.'''
```

```python
# Shapiro-Wilk test for normality
shapiro_test = stats.shapiro(df[' Blood Pressure After (mmHg)'] - df[' Blood Pressure Before (mmHg)'])
print("Shapiro-Wilk test p-value:", shapiro_test.pvalue)
```
