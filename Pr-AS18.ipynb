```python
'''
Question 1
Create a function that takes a list of non-negative integers and strings and return a new list without the strings.
Examples
filter_list([1, 2, "a", "b"]) ➞ [1, 2]

filter_list([1, "a", "b", 0, 15]) ➞ [1, 0, 15]

filter_list([1, 2, "aasf", "1", "123", 123]) ➞ [1, 2, 123]

'''

class ListFilter:
    def __init__(self, lst):
        self.lst = lst
    
    def filter_list(self):
        return [x for x in self.lst if isinstance(x, int)]

filter_obj = ListFilter([1, 2, "a", "b"])
print(filter_obj.filter_list())

filter_obj = ListFilter([1, "a", "b", 0, 15])
print(filter_obj.filter_list())

filter_obj = ListFilter([1, 2, "aasf", "1", "123", 123])
print(filter_obj.filter_list())
```

```python
'''
Question 2
The "Reverser" takes a string as input and returns that string in reverse order, with the opposite case.
Examples
reverse("Hello World") ➞ "DLROw OLLEh"

reverse("ReVeRsE") ➞ "eSrEvEr"

reverse("Radar") ➞ "RADAr"
'''

def reverse(text):
    return text[::-1].swapcase()

print(reverse("Hello World"))
print(reverse("ReVeRsE"))
print(reverse("Radar"))
class StringReverser:
    def __init__(self, string):
        self.string = string
    
    def reverse(self):
        return self.string[::-1].swapcase()

reverse_obj = StringReverser("Hello World")
print(reverse_obj.reverse())

reverse_obj = StringReverser("ReVeRsE")
print(reverse_obj.reverse())

reverse_obj = StringReverser("Radar")
print(reverse_obj.reverse())

```

```python
'''
Question 3
You can assign variables from lists like this:
lst = [1, 2, 3, 4, 5, 6]
first = lst[0]
middle = lst[1:-1]
last = lst[-1]
print(first) ➞ outputs 1
print(middle) ➞ outputs [2, 3, 4, 5]
print(last) ➞ outputs 6
With Python 3, you can assign variables from lists in a much more succinct way. Create
variables first, middle and last from the given list using destructuring assignment
(check the Resources tab for some examples), where:
first ➞ 1
middle ➞ [2, 3, 4, 5]
last ➞ 6

Your task is to unpack the list writeyourcodehere into three variables, being first,
middle, and last, with middle being everything in between the first and last element. Then
print all three variables.'''


class ListUnpacker:
    def __init__(self, lst):
        self.lst = lst
    
    def unpack_list(self):
        first, *middle, last = self.lst
        return first, middle, last

unpack_obj = ListUnpacker([1, 2, 3, 4, 5, 6])
first, middle, last = unpack_obj.unpack_list()
print(first)
print(middle)
print(last)
```

```python
'''
Question 4
Write a function that calculates the factorial of a number recursively.
Examples
factorial(5) ➞ 120

factorial(3) ➞ 6

factorial(1) ➞ 1

factorial(0) ➞ 1
'''

class FactorialCalculator:
    def factorial(self, n):
        if n == 0 or n == 1:
            return 1
        else:
            return n * self.factorial(n-1)

factorial_obj = FactorialCalculator()
print(factorial_obj.factorial(5))
print(factorial_obj.factorial(3))
print(factorial_obj.factorial(1))
print(factorial_obj.factorial(0))
```

```python
'''
Question 5
Write a function that moves all elements of one type to the end of the list.
Examples
move_to_end([1, 3, 2, 4, 4, 1], 1) ➞ [3, 2, 4, 4, 1, 1]
# Move all the 1s to the end of the array.

move_to_end([7, 8, 9, 1, 2, 3, 4], 9) ➞ [7, 8, 1, 2, 3, 4, 9]

move_to_end(["a", "a", "a", "b"], "a") ➞ ["b", "a", "a", "a"]

'''
class ElementMover:
    def move_to_end(self, lst, element):
        return [x for x in lst if x != element] + [x for x in lst if x == element]

move_obj = ElementMover()
print(move_obj.move_to_end([1, 3, 2, 4, 4, 1], 1))
print(move_obj.move_to_end([7, 8, 9, 1, 2, 3, 4], 9))
print(move_obj.move_to_end(["a", "a", "a", "b"], "a"))
```
